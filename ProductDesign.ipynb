```python
from diffusers import DiffusionPipeline

pipeline = DiffusionPipeline.from_pretrained("stablediffusionapi/product-design")
pipeline.to('cuda')
```

```python
prompt = "Design a water bottle with a sleek, cylindrical shape and a stainless steel screw cap. The bottle should have a minimalist label area and be easy to grip"

# Generate the image
generator = pipeline(prompt=prompt, num_inference_steps=30, guidance_scale=7.5, generator=None, seed=10)

# Display the image
image = generator.images[0]
image
```
